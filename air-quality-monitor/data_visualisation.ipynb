```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('air_data.csv', index_col=0)
df['Window open'] = df['Window open'].replace({True: 100, False: 0})
```

```python
df['Mean AQI'] = df['AQI'].rolling(window=12).mean()
df['Mean CO2'] = df['CO2'].rolling(window=12).mean()
df['Mean Outside AQI'] = df['Outside AQI'].rolling(window=12).mean()
df['Mean TVOC'] = df['TVOC'].rolling(window=12).mean()
```

```python
px.line(df)
```
